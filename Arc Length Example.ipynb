```python
import random
import numpy as np
import matplotlib.pyplot as plt

def length(x1,y1,x2,y2):
    return ((x2-x1)**2+(y2-y1)**2)**.5

def func(x):
    return (-630*np.cosh(x/239.2)+1260)

N = 10000
a = -315
b = 315
i = (b-a)/N

xvalues = np.arange(a,b,i)
yvalues = [func(i) for i in xvalues]

points = []
lengths = []

for i in range(N-1):
    points.append([xvalues[i],func(xvalues[i]),xvalues[i+1],func(xvalues[i+1])])
    
for i in points:
    lengths.append(length(*i))


plt.plot(xvalues,yvalues, color = "blue",linestyle = "-")
plt.title("Gateway Arch Function (ft)")
plt.show()


print("The approximate length of the gateway arch is %.2f ft" % sum(lengths))
```
